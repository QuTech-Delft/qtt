# Example MongoDB storage interface

With the storage interface we can store documents into a database.

```python
import qilib.utils.storage

storage = qilib.utils.storage.StorageMongoDb('test')

storage.save_data({'string': 'hello', 'float': 1.7}, ['stored_document'])
```

A GUI for the database is [MongoDB Compass](https://www.mongodb.com/products/compass). To search for documents in the database, enter a query in the `FILTER` field. For example to search for all documents with tag containing the text `settings` use the query `{tag: /settings/}`.
