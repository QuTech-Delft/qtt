# Fit scan of Ohmic contact

Create sample data: a linear relation between the bias voltage and the current trough the device with some random noise.

```python
%matplotlib inline
import qcodes.tests.data_mocks
import numpy as np
import qtt
from qtt.algorithms.ohmic import fitOhmic

ds = qcodes.tests.data_mocks.DataSet1D()
x = np.arange(-200, 200)
y = 1e-10 * (x + 50 + 20 * np.random.rand(x.size))
ds = qtt.data.makeDataSet1Dplain('gate', x, xunit='mV', yname='current', y=y)
_ = qcodes.MatPlot(ds.default_parameter_array())
```

Fit the data with a linear function and plot the results.

```python
r = fitOhmic(ds, fig=300, gainx=1e-6, gainy=1)
```
