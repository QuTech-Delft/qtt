We determine the pinch-off value of a 1D gate scan. 

```python
import os
import pprint
import qtt

exampledir=os.path.join(qtt.__path__[0], 'exampledata')
dataset=qtt.data.load_dataset(os.path.join(exampledir, 'pinchoff', '2018-07-17', '12-49-22_qtt_scan1D' ))
```

```python
result=qtt.algorithms.gatesweep.analyseGateSweep(dd=dataset)
qtt.algorithms.gatesweep.plot_pinchoff(result, ds=dataset, fig=10)
```

```python
pprint.pprint(result)
```
